```python
from matplotlib import pyplot as plt

tv_shows = ["Breaking Bad", "Daredevil", "Peaky Blinders"]

num_oscars = [20, 3, 6]

plt.bar(range(len(tv_shows)), num_oscars)

plt.title('My favorite tv shows')

plt.ylabel('# of academy awards')

plt.xticks(range(len(tv_shows)), tv_shows)

plt.show()
```
