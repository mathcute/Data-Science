##– Sejam os dados da quantidade de visitantes de website lançado recentemente. Deseja-se estimar um gráfico de dispersão, considerando os 20 primeiros dias desde o lançamento do website (1-20) e que o primeiro dia ele recebeu 20 visitantes e cada dia +20 visitantes do que o dia anterior (Dia 1: 20, Dia 2: 40, Dia 3: 60, ...).Use a Regressão Linear Simples.

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(1, 20, 1, dtype = float)
y = np.arange(20, 400, 20, dtype = float)

coeficiente_angular, intercepto = np.polyfit(x, y, 1)

y_pred = coeficiente_angular * x + intercepto

# grafico de dados
plt.scatter(x, y, label = 'Dados originais')
plt.plot(x, y_pred, color = 'black', linewidth = 2, label = 'Linha de Regressao')
plt.xlabel('Dias desde o lançamento do website')
plt.ylabel('Numeros de visitantes do website')
plt.legend()
plt.title('Visitantes do Website')
plt.grid(True)
plt.show()
```
